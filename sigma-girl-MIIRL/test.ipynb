```python
n_games = 50
sleep_time = 0
```

```python
for x in range(10):
    env = GridWorld(shape=(5,5), direction=direction, rew_weights=[1,15,1], fail_prob=fail_prob, goal = None, start=None)
    env_mdp = env.get_mdp()
    pi,V = env_mdp.value_iteration(max_iter=1000)
    for i in range(n_games):    
        game_data = []
        """"
        game_data = 
        [ 
            [state, action],
            [[row, column]], action]
        ]
        where state = [row,column]
        where action = 0,1,2,3
        """

        state = env.reset()
        save_state = np.array(env._intToCouple(state)[::-1])
        save_state[0] = abs(save_state[0] - env.H + 1)

        env._render()
        done = False
        while not done:
            time.sleep(sleep_time)
            action = pi[state]
            game_data.append([save_state, action])

            state, reward, done, _ = env.step(action)

            env._render()
            save_state = np.array(env._intToCouple(state)[::-1])
            save_state[0] = abs(save_state[0] - env.H + 1)

            print(save_state)
        game_data.append([save_state, action])
        game_data.append(game_data)
    file_name = 'policy-{}-games-{}-slip{}'.format(direction,n_games, int(fail_prob*100))
    games_array = np.array([np.array(x) for x in game_data])
    np.save(os.path.join('data','gridworld_discrete',file_name), games_array)
    fail_prob = fail_prob + 0.1
```

```python
for i in range(numSlips):
  for game in data_down[i]:
    states = game[:, 0]
    state_avg = np.zeros(2)
    for state in states:
      state_avg += state
    state_avg /= len(states)
    summarized_down_data.append(state_avg)
  for game in data_center[i]:
    states = game[:, 0]
    state_avg = np.zeros(2)
    for state in states:
      state_avg += state
    state_avg /= len(states)
    summarized_center_data.append(state_avg)
  for game in data_up[i]:
    states = game[:, 0]
    state_avg = np.zeros(2)
    for state in states:
      state_avg += state
    state_avg /= len(states)
    summarized_up_data.append(state_avg)
```

```python
for i in range(numSlips):
    summarized_down_data[i] = np.array(summarized_down_data[i])
    summarized_up_data[i] = np.array(summarized_up_data[i])
    summarized_center_data[i] = np.array(summarized_center_data[i])
summarized_down_data = np.array(summarized_down_data)
summarized_up_data = np.array(summarized_up_data)
summarized_center_data = np.array(summarized_center_data)
```

```python
from sklearn.cluster import KMeans
```

```python
combined_data = []
for i in range(numSlips):
    combined_data[i] = np.vstack([summarized_center_data[i], summarized_down_data[i], summarized_up_data[i]])
```

```python
kmeans = []
for i in range(numSlips):
    kmeans[i] = KMeans(n_clusters=3).fit(combined_data[i])
```
